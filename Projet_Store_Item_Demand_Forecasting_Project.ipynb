```
y_target = train_new.sales
col = [i for i in data_combine.columns if i not in ['date','id','sales','sale_moving_average_7days']]
```

```

X_train, X_test, y_train, y_test = train_test_split( train_new[col] ,train_new.sales, test_size=0.15, random_state=42)
```

```
y_train.head()
```

```
import xgboost as xgb
from sklearn.metrics import mean_squared_error

model_sets = []
for max_depth in range(4, 17, 3):
    xgb_model = xgb.XGBRegressor(max_depth=max_depth, min_child_weight=1, eval_metric='mae')
    xgb_model.fit(X_train, y_train.values)
    model_sets.append(xgb_model)

    y_train_pred_xgb = xgb_model.predict(X_train)
    y_test_pred_xgb = xgb_model.predict(X_test)
    train_smape = np.mean(np.abs(y_train.values - y_train_pred_xgb) / (np.abs(y_train.values) + np.abs(y_train_pred_xgb))) * 100
    test_smape = np.mean(np.abs(y_test.values - y_test_pred_xgb) / (np.abs(y_test.values) + np.abs(y_test_pred_xgb))) * 100
    print('SMAPE error: max_depth =', max_depth, ', train:', train_smape, 'test:', test_smape)
    print('MSE train:', mean_squared_error(y_train, y_train_pred_xgb), 'test:', mean_squared_error(y_test, y_test_pred_xgb))
```

```
def plot_xgb_importances(model, plot=False, num=10):
    importance = model.get_booster().get_score(importance_type='gain')
    importance = sorted(importance.items(), key=lambda kv: kv[1], reverse=True)
    imp_df = pd.DataFrame(importance[:num], columns=['feature', 'gain'])
    if plot:
        plt.figure(figsize=(15, 7))
        sns.set(font_scale=1)
        sns.barplot(x="gain", y="feature", data=imp_df)
        plt.title('Feature Importances')
        plt.tight_layout()
        plt.show()
    else:
        print(imp_df)
    return imp_df
```
